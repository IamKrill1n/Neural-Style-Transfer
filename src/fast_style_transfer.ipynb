```python
from nst_torch.trainer import FastStyleTransferTrainer
from nst_torch.inference import FastStyleTransfer
from nst_torch.utils import image_loader, imshow
```

```python
style_image_path = 'D:\\dev\\project\\Neural-Style-Transfer\\style_images\\sketch.jpg'
content_image_path = 'D:\\dev\\project\\Neural-Style-Transfer\\wikiart'
```

```python
trainer = FastStyleTransferTrainer()
trainer.train(style_image_path=style_image_path, content_image_path=content_image_path, checkpoint_path='D:\\dev\\project\\Neural-Style-Transfer\\checkpoints\\', batch_size = 8, epochs = 3)
```

```python
content_image = image_loader('D:\\dev\\project\\Neural-Style-Transfer\\content_images\\leopard.jpg')
```

```python
imshow(content_image)
```

```python
stylized_image = trainer.stylize(content_image=content_image)
```

```python
imshow(stylized_image)
```
