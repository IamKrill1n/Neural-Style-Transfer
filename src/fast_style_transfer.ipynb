```python
from nst_torch.trainer import FastStyleTransferTrainer
from nst_torch.inference import FastStyleTransfer
from nst_torch.utils import image_loader, imshow
```

```python
style_image_path = 'D:\\dev\\project\\Neural-Style-Transfer\\style_images\\starry_night.jpg'
content_image_path = 'D:\\dev\\project\\Neural-Style-Transfer\\content_images'
```

```python
trainer = FastStyleTransferTrainer()
trainer.train(style_image_path=style_image_path, content_image_path=content_image_path, model_path= 'D:\\dev\\project\\Neural-Style-Transfer\\model', batch_size = 4, epochs = 2)
```

```python
# content_image = image_loader('D:\\dev\\project\\Neural-Style-Transfer\\data\wikiart\\000000000008.jpg')
content_image = image_loader('D:\\dev\\project\\Neural-Style-Transfer\\content_images\\random\\leopard.jpg')
```

```python
imshow(content_image)
```

```python
trainer.transformer.to('cuda')
stylized_image = trainer.stylize(content_image=content_image)
```

```python
imshow(stylized_image)
```
