# Example Usage

```python
from nst_torch.trainer import FastStyleTransferTrainer
from nst_torch.inference import FastStyleTransfer
from nst_torch.utils import image_loader, imshow, save_image, image_unloader
import os
```

## Training

Define input data for training

```python
style_image_path = 'D:\\dev\\project\\Neural-Style-Transfer\\style_images\\starry_night.jpg'
content_image_path = 'D:\\dev\\project\\Neural-Style-Transfer\\content_images'
```

```python
trainer = FastStyleTransferTrainer()
trainer.train(style_image_path=style_image_path, content_image_path=content_image_path, model_path= 'D:\\dev\\project\\Neural-Style-Transfer\\model', batch_size = 4, epochs = 2)
```

```python
# content_image = image_loader('D:\\dev\\project\\Neural-Style-Transfer\\data\wikiart\\000000000008.jpg')
content_image = image_loader('D:\\dev\\project\\Neural-Style-Transfer\\content_images\\random\\girl.jpg')
```

```python
imshow(content_image)
```

```python
trainer.stylize(content_image)
```

## Inference

```python
# Load Model
model = FastStyleTransfer(model_path = 'D:\\dev\\project\\Neural-Style-Transfer\\model\\starry_night.model')
# model = FastStyleTransfer(model_path = 'D:\\dev\\project\\Neural-Style-Transfer\\model\\mosaic_4e5_e2.pth')
```

```python
stylized_img = model.stylize(content_image=content_image)
```

```python
output_path = 'D:\\dev\\project\\Neural-Style-Transfer\\output_images'
stylized_img = image_unloader(stylized_img)
stylized_img.save(os.path.join(output_path, f'girl_styled_as_starry_night.png'))
```
